UNDER SAMPLING


Build a sample dataset containing similar distribution of normal transaction and fraudulent transaction

No of fradulent transaction - 85

```python
legit_sample = legit.sample(n=85)
```

Concatenating two data frame

```python
new_dataset = pd.concat([legit_sample, fraud] , axis=0 )

#values will be added row wise =0
#values will be added column wise =1
```

```python
new_dataset.head() #values will be added randomly
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts() #uniformly distributed data
```

```python
new_dataset.groupby('Class').mean()
```

Spliting the data into features and targets (x and y)

```python
X= new_dataset.drop(columns = 'Class' , axis=1)
Y= new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

SPLIT THE DATA INTO TRAINING AND TESTING DATA

80- training data
20 - testing data

80% of data in X train

20% data in X test

correspondng of X train will be stored in Y train


similarly of X test wll be stored in Y test


```python
X_train , X_test , Y_train , Y_test = train_test_split(X, Y,test_size=0.2 , stratify=Y,  random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

MODEL TRAINING
using logistic Regression for Binary classification

```python
#already imported
model = LogisticRegression()
```

```python
#training the logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)  # Increase max_iter
model.fit(X_train, Y_train)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', max_iter=1000)  # Change solver and increase max_iter
model.fit(X_train, Y_train)
```

MODEL EVALUATION

 Accuracy Score


Now that our model has learn from the data SO we will only give X train value  and it will predict the value in X train


```python

  X_train_prediction = model.predict(X_train)
  training_data_accuracy = accuracy_score(X_prediction, Y_train)
```

Accuracy Score on Training Data -

```python
from sklearn.metrics import accuracy_score

# Predict on training data
X_train_prediction = model.predict(X_train)

# Calculate accuracy on training data
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)

print(f'Training Data Accuracy: {training_data_accuracy}')
```

```python
# Accuracy Score on Test Data is
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test Data :  ', test_data_accuracy)
```
